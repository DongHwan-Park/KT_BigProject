# 1. 데이터 생성

```python
# 라이브러리
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
data_origin = pd.read_csv('아파트(전월세)_실거래가_2022_중구.csv', encoding='cp949')
data = data_origin.copy()

data.tail()
```

```python
data.reset_index(drop= True, inplace = True)
data.isna().sum()
```

```python
del_cols = ['종전계약 보증금 (만원)','종전계약 월세 (만원)','계약기간','건축년도','계약구분','갱신요구권 사용']
data.drop(axis = 1 , columns=del_cols, inplace=True)
data.tail()
```

```python
data = data.loc[data['전월세구분'] == '월세']
data.reset_index(drop= True, inplace=True, )
data.tail()
```

```python
# 카카오API를 사용하여 주소->좌표 변환
# blog 
# https://parkgihyeon.github.io/project/geocoding-api/#%EA%B0%9C%EC%9D%B8-key-%EB%B0%9C%EA%B8%89%EB%B0%9B%EA%B8%B0
import requests, json

def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
    # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
    headers = {"Authorization": "KakaoAK 2550728cf88fe76f21185b7b79eddac8"}
    
    try:      
        api_json = json.loads(str(requests.get(url,headers=headers).text))
        address = api_json['documents'][0]['address']
        crd = {"lat": str(address['y']), "lng": str(address['x'])}
        address_name = address['address_name']
        return crd
    except:
        crd = {"lat": 0 , "lng" : 0 }
        return crd
```

```python
get_location('서운로 197')
```

```python
# 도로명 -> 위경도 칼럼 추가  : 무한 루프

data['위경도'] = data['도로명'].apply(lambda x : get_location(x))
data['위도'] = data['위경도'].apply(lambda x : x['lat'])
data['경도'] = data['위경도'].apply(lambda x : x['lng'])
data.drop(columns='위경도', inplace=True)

data.tail()
```

```python
# 총 7개만 0,0 으로 위경도 출력
tmp = data.loc[data['위도']== 0]
tmp
```

```python
nulls = data[data['위도']==0]
null_index = list(nulls.index)
null_index
```

```python
# 해당 index 부분만 지번번호 사용하여 주소정보 재추출( rough 하게 시군구 주소에서만 따옴 )
for i in null_index:
    tmp = get_location(data['시군구'][i])
    data['위도'][i] = tmp['lat']
    data['경도'][i] = tmp['lng']

# 아직 lat, lng 0,0 인 곳이 있는지 확인
nulls = data[data['위도']==0]
nulls
```

```python
data.tail()
```

```python
tmp = data.copy()
```

```python
# 최종 필요한feature들만 남기기

del_cols = ['시군구','번지','본번','부번','전월세구분','도로명']
data.drop(axis=1 , columns=del_cols, inplace=True)
data
```

```python
# 데이터 저장
data.to_csv('2022_중구_월세.csv', index=False, encoding="utf-8-sig")
```
