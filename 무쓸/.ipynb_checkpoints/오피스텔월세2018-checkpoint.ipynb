```python
# 카카오API를 사용하여 주소->좌표 변환
# blog 
# https://parkgihyeon.github.io/project/geocoding-api/#%EA%B0%9C%EC%9D%B8-key-%EB%B0%9C%EA%B8%89%EB%B0%9B%EA%B8%B0
import requests, json

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK 2550728cf88fe76f21185b7b79eddac8"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd

crd = get_location("중랑천로14길 58")
print(crd)
```

# 1.서울_월세_2018

```python
path = './데이터/오피스텔월세서울/국토부_오피스텔월세_2018.csv'

data_origin = pd.read_csv(path, encoding='cp949')
data_origin.tail()
```

```python
data = data_origin.copy()
```

```python
# data['전월세구분'].value_counts()
```

```python
# 월세만 남기기
data = data.loc[data['전월세구분'] == '월세']

# index 초기화
data.reset_index(drop=True, inplace = True)
data.tail()
```

```python
data.isna().sum()
```

```python
tmp = data.head()
```

```python
tmp['위도'] = tmp['도로명'].apply(lambda x : get_location(x)['lat'])
tmp.head()
```

```python
# 도로명 -> 위경도 칼럼 추가
try :
    data['위도'] = data['도로명'].apply(lambda x : get_location(x)['lat'])
    data['경도'] = data['도로명'].apply(lambda x : get_location(x)['lng'])
except:
    data['위도'] = 0
    data['경도'] = 0
data.tail()


# t = data.tail()
# t['위도'] = t['도로명'].apply(lambda x : get_location(x)['lat'])
# t['경도'] = t['도로명'].apply(lambda x : get_location(x)['lng'])
# t
```

```python
def geocoding_2(address):
    try:
        geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
        geo = geolocoder.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0,0]
```

# 데이터 알아보기

```python
data.isna().sum()
```

```python
def get_addrs(x):
    x1 = x.split(' ')
    return " ".join(x1)
    
# 주소 전처리
data['도로명주소'] = data['도로명'].apply(get_addrs)
# data['지번주소'] = data['소재지_지번'].apply(get_addrs)
data.tail()
```

```python
geo_local = Nominatim(user_agent='South Korea')

#위도 반환 함수
def geocoding_lat(address):
    try:
        geo = geo_local.geocode(address)
        return geo.latitude

    except:
        return 0

#경도 반환 함수
def geocoding_lon(address):
    try:
        geo = geo_local.geocode(address)
        return geo.longitude

    except:
        return 0
        
# 변환
data['위도'] = data['도로명'].apply(geocoding_lat)
data['경도'] = data['도로명'].apply(geocoding_lon)
data.tail()
```
